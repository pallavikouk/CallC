```python
import pandas as pd
import numpy as np
```

```python
# Comments Data Cleaning 
def read_data_from_csv():
    comments=pd.read_csv('comments.csv')
    return comments
```

```python
read_data_from_csv()
```

```python
def data_cleaning():
    comments=read_data_from_csv()
    comments.set_index('id',inplace=True)
    comments.drop(['posted date','emoji used','Hashtags used count'],axis=1,inplace=True)
    comments.rename(columns={
        'comment':'comment_text',
        'User  id':'user_id',
        'Photo id':'photo_id',
        'created Timestamp':'created_at'
    },inplace=True)
    comments.to_csv('comments_cleaned.csv')
    return comments
```

```python
data_cleaning()
```

```python
def task_runner():
    data_cleaning()
```

```python
task_runner()
```
